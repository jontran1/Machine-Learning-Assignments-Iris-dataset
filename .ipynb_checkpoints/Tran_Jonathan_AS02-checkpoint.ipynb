```python
from sklearn import datasets
import numpy as np
```

```python
# Loads the iris data into jupyer's memory.
iris = datasets.load_iris()
y = iris.target
```

```python
# To evaluate how well a trained model performs on unseen data, 
# we will further split the dataset into separate training and 
# test datasets
from sklearn.model_selection import train_test_split
```

```python
# Here, we will standardize the features using the StandardScaler class 
# from scikit-learn's preprocessing module
from sklearn.preprocessing import StandardScaler
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

def plot_decision_regions(X, y, classifier, test_idx=None,
    resolution=0.02):
    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
    np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=colors[idx],
                    marker=markers[idx], label=cl,
                    edgecolor='black')
        
    # highlight test samples
    if test_idx:
        # plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(X_test[:, 0], X_test[:, 1],
                    c='', edgecolor='black', alpha=1.0,
                    linewidth=1, marker='o',
                    s=100, label='test set')
```

```python
# Training a logistic regression model with scikitlearn
from sklearn.linear_model import LogisticRegression
```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
from sklearn.utils.testing import ignore_warnings
from sklearn.exceptions import ConvergenceWarning
from sklearn.metrics import log_loss
from sklearn.metrics import accuracy_score


@ignore_warnings(category=ConvergenceWarning)
def train_Ir_model(X, y, feature_names, max_iter_n, C, show_graph):
    """Training the linear regression model and graphing the data
    Parameters. Takes the X array and splits the data into training 
    data and testing data. The testing data is then used to show 
    accuracy. Graph will not display if feature_names is not passed into
    the function. 
    ----------
    X : array-like, shape = [n_samples, n_features]
        n_samples the number of samples
        n_features the number of features for each sample.
        
    y : array-like, shape = [n_samples]
        n_sample target values.
        
    feature_names : array shape = [n_feature_names]
        A list of feature names
        
    max_iter_n : max number of iterations for training the model.
    
    C : Float Inverse of regularization strength; must be a positive float. 
        Like in support vector machines, smaller values specify stronger regularization.
        
    show_graph : boolean if true graph will be displayed.
        
    Return 
        (float) accuracy of logistic model.
        (float) logistic loss value.
    
    """
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=1, stratify=y)

    sc = StandardScaler()
    sc.fit(X_train)

    X_train_std = sc.transform(X_train)
    X_test_std = sc.transform(X_test)

    X_combined_std = np.vstack((X_train_std, X_test_std))
    y_combined = np.hstack((y_train, y_test))

    lr = LogisticRegression(C=C, random_state=1, max_iter=max_iter_n)
    lr.fit(X_train_std, y_train)
    
    if show_graph and feature_names is not None:
        plot_decision_regions(X_combined_std, 
                              y_combined, 
                              classifier=lr, 
                              test_idx=range(105, 150))

        plt.xlabel(feature_names[0] + ' [standardized]')
        plt.ylabel(feature_names[1] + ' [standardized]')
        plt.legend(loc='upper left')
        plt.show()

    y_prediction = lr.predict_proba(X_test_std)
    y_pred = y_prediction.argmax(axis=1)
    loss = log_loss(y_test, y_prediction, normalize=False)
    return accuracy_score(y_test, y_pred), loss;
```

```python
# Exmaple of the logistic model using 10 iterations to train the model.
# The graph shows the 3 classifications seperated by color. 
feature_names = [' Sepal Length', ' Sepal Width', ' Petal Length', ' Petal Width']
columns = ['names_of_features', 'feature_names', 'number_of_iterations', 'accuracy', 'logistic loss']

X = iris.data[:, [0, 1]]
accuracy, loss = train_Ir_model(X,y,["Sepal Length", "Sepal Width"], 10, 100.0, True)
print('Accuracy: %.2f' % accuracy)
print('Accuracy: %.5f' % loss)
```

```python
def graph_accuracy_and_loss(n_iterations, accuracy, loss):
    """This function will plot and display two scatter point graphs
    ---------
    n_iterations : list, shape = [n_iterations]
    n_iterations should only contain positive integers. 
    
    accuracy : list, shape = [n_accuracy]
    
    list : list, shape = [n_loss]    
    """
    
    plt.scatter(n_iterations, loss, marker='x')
    plt.xlabel('n iterations')
    plt.ylabel('loss')
    plt.show()
    plt.scatter(n_iterations, accuracy, marker='x')
    plt.xlabel('n iterations')
    plt.ylabel('accuracy')
    plt.show()
```

```python
# Converges based on accuracy and loss
# All one case using all features..
def run_simulation_and_graph(x, y, names_of_features, n_simulations, C, logistic_graph, graph_accuracy_loss):
    
    n_iter = [n for n in range(1, n_simulations)]
    accuracy_list = []
    loss = []
    X = x
    save_convergence = True

    for n in range(1,n_simulations):
        percent_accuracy,logistic_loss = train_Ir_model(X,y,names_of_features, n, C, logistic_graph)
        loss.append(logistic_loss) 
        accuracy_list.append(percent_accuracy)
        # if new-old < 0.5 then save these values for convergence. 
        if n > 1 and abs(loss[n-1]-loss[n-2]) < 0.5 and save_convergence:
            save_convergence = False
            coverg_loss = loss[n-1]
            coverg_n_iter = n
            converg_percent = accuracy_list[n-1]
    if graph_accuracy_loss:
        graph_accuracy_and_loss(n_iter, accuracy_list, loss)
    return [len(names_of_features), ', '.join(names_of_features), coverg_n_iter, converg_percent, coverg_loss]
```

```python
import pandas as pd

columns = ['names_of_features', 'feature_names', 'number_of_iterations', 'accuracy', 'logistic_loss']

# This data frame was created from the interations performed above. 
# If the abs(logistic_loss old - logistic_loss new) < 0.5 then we save the current number of iterations,
# percentage accuracy and logistic loss. There is no reason for 0.5, i simply picked it because it seem
# small enough. In actuality if the difference was small enough we would stop training the model to save
# computation power.
#

# C parameter is 100 by default
convergence_df = pd.DataFrame(np.array([case_0,case_1,case_2,case_3,case_4,case_5,case_6,case_7,case_8,case_9,case_10]),
                 columns=columns)

convergence_df

# Findings
# ------------------------------------------------------------------------------------------------------------
# It seems the best features to use is Sepal Width, Petal Width, seeing as how it converges to an answer 
# with only 7 iterations with an accurancy of % 100 based on the training data ofcourse. The combination
# of features is Sepal Length, Sepal Width. However, it did converge quickly after 3 iterations its 
# accuracy is low compare to the others with a percentage of 75%. If we were to use all features at the same time
# the accuracy comes to 97 percent with a logistic loss of 7.9. It only needs 8 iterations to converge. However,
# comapre to Sepal Width, Petal Width with a 100% accuracy, its logistic loss is lower, althrough not by much. 
# Right off the bat you would think Sepal Width, Petal Width is the way to go with its 100% accuracy but I 
# believe if you were to give it a different testing set the accuracy would differ. Also I played around with the code
# and changed the random state to another value and saw that it increase the accuracy in some cases. Based on the chart
# I believe the best case would be to go with one of the two features with an accuracy higher than 97% to save 
# computation power rather than the four or six feature cases. 
# ------------------------------------------------------------------------------------------------------------
```

```python
columns = ['names_of_features', 'feature_names', 'number_of_iterations', 'accuracy', 'logistic_loss', 'inverse_of_regularization_strength(C)']

X = iris.data[:, [0, 1]]
cases = []
for n in range(1,600, 50):
    case = run_simulation_and_graph(X, y, ["Sepal Length", "Sepal Width"], 10, n, False, False)
    case.append(n)
    cases.append(case)
    

df = pd.DataFrame(np.array(cases),
                 columns=columns)

df    

# Findings
# ------------------------------------------------------------------------------------------------------------
# Playing around with the regularization strength (C parameter of logistic regression) I found that increasing
# the regularization strength does increase accuracy all except for case Sepal Length, Sepal Width which stays at 
# 75%. What I found was that logistic regression was able to converge faster if the regularization strength was 
# very low i.e 1. However, this doesn't mean its very accurate. I found that increasing the regularization strength 
# has a the "diminishing returns" problem. Increasing the strength pass a certain point will only cause minor changes 
# in accuracy and logistic loss and its not worth the computational power to do so. The power is finding out what
# C parameter to use, finding out when it converges, finding out when to stop training the model. 
# ------------------------------------------------------------------------------------------------------------
# Below are my charts for different inverse_of_regularization_strengths or (C parameters)
```
